```python
Q1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.
Line Plot:

Use: To visualize trends over time or a continuous variable.
Example: Plotting time series data, such as stock prices over a period of time.
Scatter Plot:

Use: To show the relationship between two continuous variables.
Example: Visualizing the correlation between height and weight.
Box Plot:

Use: To display the distribution of a dataset based on a five-number summary (minimum, first quartile, median, third quartile, and maximum).
Example: Comparing the distribution of test scores between different groups.
Histogram:

Use: To show the distribution of a single continuous variable.
Example: Displaying the frequency of different ranges of values in a dataset, such as the distribution of ages.
Heatmap:

Use: To display data in matrix form, with values represented by colors.
Example: Showing the correlation between different variables in a dataset.
Q2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
fmri = sns.load_dataset("fmri")

# Plot the line plot
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event", style="region")
plt.title('FMRI Signal Over Time for Different Events and Regions')
plt.xlabel('Timepoint')
plt.ylabel('Signal')
plt.show()
Q3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.
# Load the dataset
titanic = sns.load_dataset("titanic")

# Plot the box plots
plt.figure(figsize=(14, 6))

# Box plot for 'age'
plt.subplot(1, 2, 1)
sns.boxplot(data=titanic, x='pclass', y='age')
plt.title('Age Distribution by Pclass')

# Box plot for 'fare'
plt.subplot(1, 2, 2)
sns.boxplot(data=titanic, x='pclass', y='fare')
plt.title('Fare Distribution by Pclass')

plt.show()
Q4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.
# Load the dataset
diamonds = sns.load_dataset("diamonds")

# Plot the histogram
plt.figure(figsize=(10, 6))
sns.histplot(data=diamonds, x='price', hue='cut', multiple='stack')
plt.title('Distribution of Diamond Prices by Cut')
plt.xlabel('Price')
plt.ylabel('Count')
plt.show()
Q5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset
iris = sns.load_dataset("iris")

# Plot the pair plot
sns.pairplot(data=iris, hue="species")
plt.suptitle('Pair Plot of Iris Dataset', y=1.02)
plt.show()
Q6: Use the "flights" dataset from seaborn to plot a heatmap.

# Load the dataset
flights = sns.load_dataset("flights")

# Pivot the dataset to create a matrix
flights_pivot = flights.pivot("month", "year", "passengers")

# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(flights_pivot, annot=True, fmt="d", cmap="YlGnBu")
plt.title('Heatmap of Flights Dataset')
plt.xlabel('Year')
plt.ylabel('Month')
plt.show()
```
