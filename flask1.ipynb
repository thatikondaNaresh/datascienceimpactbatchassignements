```python
Q1. Flask Framework and Its Advantages
Flask Framework:
Flask is a lightweight and flexible Python web framework that provides tools, libraries, and templates to help developers build web applications quickly and easily.
It is designed with simplicity and minimalism in mind, allowing developers to choose the tools and libraries they need.
Flask is often used for developing small to medium-sized web applications and APIs.
Advantages of Flask Framework:
Simplicity: Flask has a simple and easy-to-understand API, making it ideal for beginners and experienced developers alike.
Flexibility: Flask allows developers to choose their components, such as database, template engine, and more.
Built-in Development Server: Flask comes with a built-in development server and debugger, making it easy to get started with development.
Extensive Documentation: Flask has comprehensive documentation that helps developers understand its concepts and use cases effectively.
Widely Adopted: Flask has a large community of developers and extensive third-party extensions, ensuring robust support and scalability for projects.

Q2. Simple Flask Application to Display 'Hello World!!'
Here's a simple Flask application to display "Hello World!!":
from flask import Flask

# Create a Flask app instance
app = Flask(__name__)

# Define a route for the root URL
@app.route('/')
def hello_world():
    return 'Hello World!!'

# Run the Flask app
if __name__ == '__main__':
    app.run(debug=True)
Q3. App Routing in Flask
App Routing:

App routing in Flask refers to the process of mapping URLs (routes) to functions that handle those URLs.
We use app routes in Flask to define the behavior of our web application based on the URL requested by the client.
Each route specifies a URL path and the function that should be executed when that URL is accessed.
Q4. Flask Routes Example
Here's how you can create routes /welcome and / in Flask:
from flask import Flask

# Create a Flask app instance
app = Flask(__name__)

# Route for '/welcome'
@app.route('/welcome')
def welcome():
    return 'Welcome to ABC Corporation'

# Route for '/'
@app.route('/')
def company_details():
    return '''
    <h1>Company Name: ABC Corporation</h1>
    <p>Location: India</p>
    <p>Contact Detail: 999-999-9999</p>
    '''

# Run the Flask app
if __name__ == '__main__':
    app.run(debug=True)
Q5. Flask URL Building with url_for() Function
url_for() Function:

url_for() is a function provided by Flask to build URLs dynamically for a given function.
It generates a URL for the specified endpoint using the URL rules defined in the application.
This function simplifies the process of linking to specific endpoints in Flask applications.
Example:
from flask import Flask, url_for

# Create a Flask app instance
app = Flask(__name__)

# Route for '/'
@app.route('/')
def index():
    return 'Index Page'

# Route for '/user/<username>'
@app.route('/user/<username>')
def profile(username):
    return f'User {username}'

# Route for '/login'
@app.route('/login')
def login():
    return 'Login Page'

# Demonstrate url_for() usage
with app.test_request_context():
    print(url_for('index'))     # Output: '/'
    print(url_for('login'))     # Output: '/login'
    print(url_for('profile', username='John'))  # Output: '/user/John'
In this example, url_for() generates URLs for the index, login, and profile endpoints defined in the Flask application.

```
