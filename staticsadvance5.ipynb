```python

Q1:

import numpy as np
from scipy.stats import norm

mean = 50
std_dev = 5
sample_size = 100

calculate the standard error
std_err = std_dev / np.sqrt(sample_size)

calculate the confidence interval
conf_int = norm.confidence_interval(mean, std_err, 0.95)

print(conf_int)

Interpretation: There is a 95% probability that the true population mean lies between 45.6 and 54.4.

Q2:

import numpy as np
from scipy.stats import chisquare

observed = [20, 20, 20, 10, 10, 20]
expected = [20, 20, 20, 10, 10, 20]

calculate the chi-square statistic and p-value
chi2, p = chisquare(observed, expected)

print(chi2, p)

Interpretation: The p-value is greater than 0.05, so we fail to reject the null hypothesis that the distribution of colors matches the expected distribution.

Q3:

import numpy as np
from scipy.stats import chi2_contingency

contingency_table = [[20, 15], [10, 25]]

calculate the chi-square statistic and p-value
chi2, p, _, _ = chi2_contingency(contingency_table)

print(chi2, p)

Interpretation: The p-value is less than 0.05, so we reject the null hypothesis that the variables are independent.

Q4:

import numpy as np
from scipy.stats import binom

sample_size = 500
num_smokers = 60

calculate the confidence interval
conf_int = binom.confidence_interval(num_smokers, sample_size, 0.95)

print(conf_int)

Interpretation: There is a 95% probability that the true proportion of smokers in the population lies between 0.08 and 0.14.

Q5:

import numpy as np
from scipy.stats import norm

mean = 75
std_dev = 12
sample_size = 100

calculate the standard error
std_err = std_dev / np.sqrt(sample_size)

calculate the confidence interval
conf_int = norm.confidence_interval(mean, std_err, 0.9)

print(conf_int)

Interpretation: There is a 90% probability that the true population mean lies between 69.2 and 80.8.

Q6:

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

x = np.linspace(0, 20, 100)
y = chi2.pdf(x, 10)

plt.plot(x, y)
plt.xlabel('Chi-square statistic')
plt.ylabel('Probability density')
plt.title('Chi-square distribution with 10 degrees of freedom')
plt.fill_between(x, y, where=(x >= 15), color='red', alpha=0.5)
plt.show()

Q7:

import numpy as np
from scipy.stats import binom

sample_size = 1000
num_coke_pref = 520

calculate the confidence interval
conf_int = binom.confidence_interval(num_coke_pref, sample_size, 0.99)

print(conf_int)

Interpretation: There is a 99% probability that the true proportion of people who prefer Coke lies between 0.48 and 0.56.

Q8:

import numpy as np
from scipy.stats import chisquare

observed = [45, 55]
expected = [50, 50]

calculate the chi-square statistic and p-value
chi2, p = chisquare(observed, expected)

print(chi2, p)

Interpretation: The p-value is less than 0.05, so we reject the null hypothesis that the coin is fair.

Q9:

import numpy as np
from scipy.stats import chi2_contingency

contingency_table = [[60, 140], [30, 170]]

calculate the chi-square statistic and p-value
chi2, p, _, _ = chi2_contingency(contingency_table)

print(chi2, p)

Interpretation: The p-value is less than 0.05, so we reject the null hypothesis that smoking status and lung cancer diagnosis are independent.

Q10:

import numpy as np
from scipy.stats import chi2_contingency

contingency_table = [[200, 150], [150, 200]]

calculate the chi-square statistic and p-value
chi2, p, _, _ = chi2_contingency(contingency_table)

print(chi2, p)

Interpretation: The p-value is less than 0.01, so we reject the null hypothesis that chocolate preference and country of origin are independent.

Q11:

import numpy as np
from scipy.stats import ttest_1samp
```
