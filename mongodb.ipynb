```python
. MongoDB and Non-Relational Databases
MongoDB:
MongoDB is a popular open-source NoSQL database that stores data in flexible, JSON-like documents.
Non-Relational Databases: Non-relational databases (NoSQL databases) do not use the traditional tabular structure of relational databases. They offer flexible schemas and are designed to handle large volumes of unstructured or semi-structured data efficiently.
Scenarios to Prefer MongoDB:
Flexible Schema: When the schema of data is not predefined and can evolve over time.
Scalability: MongoDB is horizontally scalable, making it suitable for handling large-scale applications and big data.
Document-Oriented: If your application deals with complex data structures or hierarchical data, MongoDB's document-oriented model is beneficial.
Fast Development: For rapid prototyping or agile development where changes in data models are frequent.
Q2. Features of MongoDB
Features:
Schemaless: Documents in MongoDB do not require a predefined schema, allowing flexibility in data models.
High Performance: Supports high throughput and low latency operations due to efficient indexing and query optimization.
Horizontal Scalability: Scales horizontally by sharding, distributing data across multiple servers.
Document-Oriented: Stores data in JSON-like BSON (Binary JSON) documents, supporting nested data structures.
Rich Query Language: Supports rich queries using MongoDB Query Language (MQL), including aggregation pipeline for advanced analytics.
High Availability: Supports replica sets for automatic failover and data redundancy.
Indexing: Provides efficient indexing support to optimize query performance.
Q3. Connecting MongoDB to Python and Creating Database and Collection
import pymongo
# Establish connection to MongoDB
myclient = pymongo.MongoClient("mongodb://localhost:27017/")

# Create a database
mydb = myclient["mydatabase"]

# Create a collection
mycol = mydb["customers"]
Q4. Inserting Records and Using find() and find_one() Methods
# Insert one record
record = {"name": "John", "address": "Highway 37"}
inserted_one = mycol.insert_one(record)
print("Inserted ID:", inserted_one.inserted_id)

# Insert many records
records = [
  {"name": "Peter", "address": "Lowstreet 27"},
  {"name": "Amy", "address": "Apple st 652"},
  {"name": "Hannah", "address": "Mountain 21"},
  {"name": "Michael", "address": "Valley 345"},
  {"name": "Sandy", "address": "Ocean blvd 2"},
  {"name": "Betty", "address": "Green Grass 1"},
  {"name": "Richard", "address": "Sky st 331"},
  {"name": "Susan", "address": "One way 98"},
  {"name": "Vicky", "address": "Yellow Garden 2"},
  {"name": "Ben", "address": "Park Lane 38"},
  {"name": "William", "address": "Central st 954"},
  {"name": "Chuck", "address": "Main Road 989"},
  {"name": "Viola", "address": "Sideway 1633"}
]
inserted_many = mycol.insert_many(records)
print("Inserted IDs:", inserted_many.inserted_ids)

# Using find_one() to retrieve one record
result_one = mycol.find_one({"name": "John"})
print("Find One Result:", result_one)

# Using find() to retrieve all records
result_all = mycol.find()
for document in result_all:
    print(document)
Q5. Using find() Method to Query MongoDB
find() Method: The find() method in MongoDB is used to query documents from a collection based on specified criteria.

Example:
# Querying documents with a specific condition
results = mycol.find({"address": "Park Lane 38"})

# Printing the results
for document in results:
    print(document)
Q6. Explanation of sort() Method in MongoDB
sort() Method: The sort() method in MongoDB is used to sort the results of a query in ascending or descending order based on one or more fields.

Example:

# Sorting documents by 'name' field in ascending order
results = mycol.find().sort("name")

# Printing sorted results
for document in results:
    print(document)
Q7. Usage of delete_one(), delete_many(), and drop() in MongoDB
delete_one(filter): Deletes a single document matching the filter criteria.
mycol.delete_one({"name": "John"})
delete_many(filter): Deletes all documents matching the filter criteria.
mycol.delete_many({"address": {"$regex": "^S"}})
drop(): Deletes the entire collection, including all its documents.
mycol.drop()
These methods are used to remove documents or entire collections from the MongoDB database based on specified criteria or to delete the entire collection altogether.

```
