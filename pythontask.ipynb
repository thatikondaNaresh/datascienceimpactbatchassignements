```python
Q1. Function to Return the Product of All Numbers in a Flat List
def flatten_and_multiply(lst):
    flat_list = []
    
    def flatten(lst):
        for item in lst:
            if isinstance(item, (list, tuple, set)):
                flatten(item)
            elif isinstance(item, dict):
                for key, value in item.items():
                    if isinstance(key, (int, float)):
                        flat_list.append(key)
                    if isinstance(value, (int, float)):
                        flat_list.append(value)
                    elif isinstance(value, (list, tuple)):
                        flatten(value)
            elif isinstance(item, (int, float)):
                flat_list.append(item)
    
    flatten(lst)
    
    product = 1
    for num in flat_list:
        product *= num
    
    return product

list1 = [1, 2, 3, 4, [44, 55, 66, True], False, (34, 56, 78, 89, 34), {1, 2, 3, 3, 2, 1}, 
         {1: 34, "key2": [55, 67, 78, 89], 4: (45, 22, 61, 34)}, [56, 'data science'], 'Machine Learning']

result = flatten_and_multiply(list1)
print(result)  # Output: 78775418662400


Q2. Program for Encrypting a Message
def encrypt_message(message):
    encrypted_message = []
    for char in message.lower():
        if char.isalpha():
            encrypted_char = chr(219 - ord(char))
            encrypted_message.append(encrypted_char)
        elif char == ' ':
            encrypted_message.append('$')
        else:
            encrypted_message.append(char)
    return ''.join(encrypted_message)

input_sentence = "I want to become a Data Scientist."
encrypted_sentence = encrypt_message(input_sentence)
print(encrypted_sentence)
# Output: r$dzmg$gl$yvxlnv$z$wzgz$hxrvmgrhg.
```
