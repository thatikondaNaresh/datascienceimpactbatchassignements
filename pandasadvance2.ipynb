```python
Q1. Write a code to print the data present in the second row of the dataframe, df.
import pandas as pd

course_name = ['Data Science', 'Machine Learning', 'Big Data', 'Data Engineer']
duration = [2, 3, 6, 4]
df = pd.DataFrame(data={'course_name': course_name, 'duration': duration})

# Printing the data present in the second row
print(df.iloc[1])
Q2. What is the difference between the functions loc and iloc in pandas.DataFrame?
loc: Accesses rows and columns by labels or boolean arrays.
iloc: Accesses rows and columns by integer position.
Q3. Reindex the given dataframe using a variable reindex = [3,0,1,2] and store it in the variable new_df then find the output for both new_df.loc[2] and new_df.iloc[2].
reindex = [3, 0, 1, 2]
new_df = df.reindex(reindex)

# Output for both new_df.loc[2] and new_df.iloc[2]
print(new_df.loc[2])
print(new_df.iloc[2])
Q4. Write a code to find the following statistical measurements for the above dataframe df1:
(i) mean of each and every column present in the dataframe.
(ii) standard deviation of column, 'column_2'
import numpy as np

columns = ['column_1', 'column_2', 'column_3', 'column_4', 'column_5', 'column_6']
indices = [1, 2, 3, 4, 5, 6]

# Creating a dataframe
df1 = pd.DataFrame(np.random.rand(6, 6), columns=columns, index=indices)

# Mean of each and every column
print(df1.mean())

# Standard deviation of column 'column_2'
print(df1['column_2'].std())
Q5. Replace the data present in the second row of column, 'column_2' by a string variable then find the mean of column, 'column_2'. If you are getting errors in executing it then explain why.
# Replacing the second row of column 'column_2' with a string
df1.loc[2, 'column_2'] = 'string_value'

# Trying to find the mean of column 'column_2'
try:
    print(df1['column_2'].mean())
except Exception as e:
    print("Error:", e)
Explanation: The error occurs because the column contains a non-numeric value ('string_value'), making it impossible to calculate the mean.

Q6. What do you understand about the windows function in pandas and list the types of windows functions?
Windows functions in pandas are used to perform calculations across a set of rows that are related to the current row. Types include:

Rolling window functions (rolling)
Expanding window functions (expanding)
Exponential weighted window functions (ewm)
Q7. Write a code to print only the current month and year at the time of answering this question.
Copy code
import pandas as pd

current_date = pd.to_datetime('now')
print(current_date.strftime('%B %Y'))
Q8. Write a Python program that takes in two dates as input (in the format YYYY-MM-DD) and calculates the difference between them in days, hours, and minutes using Pandas time delta.
Copy code
import pandas as pd

def date_difference():
    date1 = input("Enter the first date (YYYY-MM-DD): ")
    date2 = input("Enter the second date (YYYY-MM-DD): ")

    date1 = pd.to_datetime(date1)
    date2 = pd.to_datetime(date2)

    delta = date2 - date1

    days = delta.days
    hours, remainder = divmod(delta.seconds, 3600)
    minutes, _ = divmod(remainder, 60)

    print(f"Difference: {days} days, {hours} hours, and {minutes} minutes")

date_difference()
Q9. Write a Python program that reads a CSV file containing categorical data and converts a specified column to a categorical data type.
import pandas as pd

def convert_to_categorical():
    file_path = input("Enter the file path: ")
    column_name = input("Enter the column name: ")
    category_order = input("Enter the category order (comma-separated): ").split(',')

    df = pd.read_csv(file_path)
    df[column_name] = pd.Categorical(df[column_name], categories=category_order, ordered=True)
    print(df.sort_values(by=[column_name]))

convert_to_categorical()
Q10. Write a Python program that reads a CSV file containing sales data for different products and visualizes the data using a stacked bar chart.
import pandas as pd
import matplotlib.pyplot as plt

def visualize_sales():
    file_path = input("Enter the file path: ")
    df = pd.read_csv(file_path)

    df.set_index('Date', inplace=True)
    df.groupby('Product').plot(kind='bar', stacked=True)

    plt.show()

visualize_sales()
Q11. Write a Python program that reads a CSV file containing student data, calculates the mean, median, and mode of the test scores, and displays the results in a table.
import pandas as pd
from scipy import stats

def analyze_student_data():
    file_path = input("Enter the file path of the CSV file containing the student data: ")
    df = pd.read_csv(file_path)

    mean = df['Test Score'].mean()
    median = df['Test Score'].median()
    mode = stats.mode(df['Test Score'])[0]

    print("+-----------+--------+")
    print("| Statistic | Value  |")
    print("+-----------+--------+")
    print(f"| Mean      | {mean:.1f}    |")
    print(f"| Median    | {median:.1f}    |")
    print(f"| Mode      | {', '.join(map(str, mode))}    |")
    print("+-----------+--------+")

analyze_student_data()
```
