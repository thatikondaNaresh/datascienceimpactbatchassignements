```python
Q1. Class and Object in Object-Oriented Programming
Class: A blueprint or prototype from which objects are created. It defines a set of attributes and methods that the objects created from the class will have.

Object: An instance of a class. It represents an individual "thing" that has the properties and behaviors defined by its class.
class Car:
    def __init__(self, brand, model, year):
        self.brand = brand
        self.model = model
        self.year = year
    
    def display_info(self):
        print(f"Brand: {self.brand}, Model: {self.model}, Year: {self.year}")

# Creating an object of the Car class
my_car = Car("Toyota", "Corolla", 2020)
my_car.display_info()  # Output: Brand: Toyota, Model: Corolla, Year: 2020

Q2. Four Pillars of OOP
Encapsulation: Bundling the data (attributes) and methods (functions) that operate on the data into a single unit or class. It restricts direct access to some of the object's components, which can prevent the accidental modification of data.

Inheritance: A mechanism where a new class inherits attributes and methods from an existing class. This helps in reusability and the creation of a hierarchical relationship between classes.

Polymorphism: The ability of different classes to be treated as instances of the same class through a common interface. It allows methods to do different things based on the object it is acting upon.

Abstraction: The concept of hiding the complex implementation details and showing only the essential features of the object. It simplifies the interaction with the object by exposing only the necessary parts.

Q3. The __init__() Function
The __init__() function is a special method in Python classes. It is the constructor method that gets called automatically when an instance (object) of the class is created. It allows the class to initialize the attributes of the object.
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display(self):
        print(f"Name: {self.name}, Age: {self.age}")

# Creating an object of the Person class
person1 = Person("Alice", 30)
person1.display()  # Output: Name: Alice, Age: 30
Q4. Why self is Used in OOP
The self keyword is used in instance methods to refer to the instance of the class. It is used to access variables that belong to the class. The first parameter of any instance method is self, which allows the method to refer to the calling instance.
class Animal:
    def __init__(self, species):
        self.species = species
    
    def make_sound(self, sound):
        print(f"The {self.species} says {sound}")

# Creating an object of the Animal class
dog = Animal("Dog")
dog.make_sound("Woof")  # Output: The Dog says Woof

Q5. Inheritance in OOP
Inheritance: A mechanism in which one class inherits attributes and methods from another class. This promotes code reuse and establishes a natural hierarchical relationship between classes.
Types of Inheritance:
Single Inheritance: A class inherits from one superclass.
python
Copy code
class Parent:
    def parent_method(self):
        print("This is the parent method")

class Child(Parent):
    def child_method(self):
        print("This is the child method")

obj = Child()
obj.parent_method()  # Output: This is the parent method
obj.child_method()   # Output: This is the child method
Multiple Inheritance: A class inherits from more than one superclass.
class Father:
    def father_method(self):
        print("This is the father method")

class Mother:
    def mother_method(self):
        print("This is the mother method")

class Child(Father, Mother):
    def child_method(self):
        print("This is the child method")

obj = Child()
obj.father_method()  # Output: This is the father method
obj.mother_method()  # Output: This is the mother method
obj.child_method()   # Output: This is the child method
Multilevel Inheritance: A class inherits from a superclass, and another class inherits from the derived class.
class Grandparent:
    def grandparent_method(self):
        print("This is the grandparent method")

class Parent(Grandparent):
    def parent_method(self):
        print("This is the parent method")

class Child(Parent):
    def child_method(self):
        print("This is the child method")

obj = Child()
obj.grandparent_method()  # Output: This is the grandparent method
obj.parent_method()       # Output: This is the parent method
obj.child_method()        # Output: This is the child method
Hierarchical Inheritance: Multiple classes inherit from a single superclass.
class Parent:
    def parent_method(self):
        print("This is the parent method")

class Child1(Parent):
    def child1_method(self):
        print("This is the child1 method")

class Child2(Parent):
    def child2_method(self):
        print("This is the child2 method")

obj1 = Child1()
obj2 = Child2()
obj1.parent_method()  # Output: This is the parent method
obj1.child1_method()  # Output: This is the child1 method
obj2.parent_method()  # Output: This is the parent method
obj2.child2_method()  # Output: This is the child2 method
Hybrid Inheritance: A combination of two or more types of inheritance.
class A:
    def method_A(self):
        print("This is method A")

class B(A):
    def method_B(self):
        print("This is method B")

class C(A):
    def method_C(self):
        print("This is method C")

class D(B, C):
    def method_D(self):
        print("This is method D")

obj = D()
obj.method_A()  # Output: This is method A
obj.method_B()  # Output: This is method B
obj.method_C()  # Output: This is method C
obj.method_D()  # Output: This is method D
```
