```python
Characteristics of Tuples:
Ordered: Elements have a defined order, and that order will not change.
Immutable: Once a tuple is created, you cannot change, add, or remove items.
Indexed: Items can be accessed by their index.
Allow Duplicates: Tuples can have items with the same value.
Can store different data types: A tuple can contain different data types (e.g., integers, strings, lists).
Immutability:
Yes, tuples are immutable. This means once a tuple is created, its elements cannot be changed, added, or removed.

Q2. Tuple Methods and Comparison with Lists
Two Tuple Methods:
count(): Returns the number of times a specified value occurs in a tuple.
example_tuple = (1, 2, 2, 3, 4, 4, 4)
count_of_4 = example_tuple.count(4)
print(count_of_4)  # Output: 3
index(): Searches the tuple for a specified value and returns the position of where it was found.
example_tuple = (1, 2, 3, 4)
index_of_3 = example_tuple.index(3)
print(index_of_3)  # Output: 2
Reason for Limited Methods:
Tuples are designed to be simple and efficient. Their immutability and fixed size mean they don't require the extensive set of methods provided for lists, which support dynamic resizing and modification.

Q3. Collection Datatypes and Removing Duplicates
Collection Datatypes without Duplicates: Sets do not allow duplicate items.
Code to Remove Duplicates using a Set:
list_with_duplicates = [1, 1, 1, 2, 1, 3, 1, 4, 2, 1, 2, 2, 2, 3, 2, 4, 3, 1, 3, 2, 3, 3, 3, 4, 4, 1, 4, 2, 4, 3, 4, 4]
unique_set = set(list_with_duplicates)
unique_list = list(unique_set)
print(unique_list)  # Output: [1, 2, 3, 4]

Q4. Union vs Update Methods for a Set
union(): Returns a new set with all items from both sets.
set1 = {1, 2, 3}
set2 = {3, 4, 5}
new_set = set1.union(set2)
print(new_set)  # Output: {1, 2, 3, 4, 5}
update(): Adds all items from one set to another.

set1 = {1, 2, 3}
set2 = {3, 4, 5}
set1.update(set2)
print(set1)  # Output: {1, 2, 3, 4, 5}

Q5. Dictionary Explanation
Dictionary: A collection of key-value pairs. Keys must be unique and immutable.
example_dict = {'name': 'Alice', 'age': 25, 'city': 'New York'}
Ordered or Unordered:
As of Python 3.7, dictionaries are ordered. This means items have a defined order, and this order will not change.

Q6. Nested Dictionary
Yes, you can create nested dictionaries.
nested_dict = {
    'person': {
        'name': 'Alice',
        'age': 25
    },
    'job': {
        'title': 'Engineer',
        'company': 'Tech Corp'
    }
}
print(nested_dict)

Q7. Using setdefault() Method
dict1 = {'language': 'Python', 'course': 'Data Science Masters'}
dict1.setdefault('topics', ['Python', 'Machine Learning', 'Deep Learning'])
print(dict1)
# Output: {'language': 'Python', 'course': 'Data Science Masters', 'topics': ['Python', 'Machine Learning', 'Deep Learning']}

Q8. Three View Objects in Dictionaries
The three view objects are keys, values, and items.
dict1 = {'Sport': 'Cricket', 'Teams': ['India', 'Australia', 'England', 'South Africa', 'Sri Lanka', 'New Zealand']}
# keys()
keys_view = dict1.keys()
print(keys_view)  # Output: dict_keys(['Sport', 'Teams'])
# values()
values_view = dict1.values()
print(values_view)  # Output: dict_values(['Cricket', ['India', 'Australia', 'England', 'South Africa', 'Sri Lanka', 'New Zealand']])
# items()
items_view = dict1.items()
print(items_view)  # Output: dict_items([('Sport', 'Cricket'), ('Teams', ['India', 'Australia', 'England', 'South Africa', 'Sri Lanka', 'New Zealand'])])

```
