```python
Q1. Python Program to Extract the Video URL of the First Five Videos
import requests
from bs4 import BeautifulSoup

def extract_video_urls():
    url = 'https://www.youtube.com/results?search_query=python+tutorial'  # Example URL
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    video_urls = []
    for i, link in enumerate(soup.find_all('a', class_='yt-uix-tile-link')[:5], start=1):
        video_url = 'https://www.youtube.com' + link['href']
        video_urls.append(video_url)
        print(f"Video {i}: {video_url}")
    
    return video_urls

# Call the function
video_urls = extract_video_urls()


Q2. Python Program to Extract the URL of Video Thumbnails of the First Five Videos
import requests
from bs4 import BeautifulSoup

def extract_thumbnail_urls():
    url = 'https://www.youtube.com/results?search_query=python+tutorial'  # Example URL
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    thumbnail_urls = []
    for i, img in enumerate(soup.find_all('img', class_='yt-img-shadow')[:5], start=1):
        thumbnail_url = img['src']
        thumbnail_urls.append(thumbnail_url)
        print(f"Thumbnail {i}: {thumbnail_url}")
    
    return thumbnail_urls

# Call the function
thumbnail_urls = extract_thumbnail_urls()


Q3. Python Program to Extract the Title of the First Five Videos
import requests
from bs4 import BeautifulSoup

def extract_video_titles():
    url = 'https://www.youtube.com/results?search_query=python+tutorial'  # Example URL
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    video_titles = []
    for i, title in enumerate(soup.find_all('a', class_='yt-uix-tile-link')[:5], start=1):
        video_title = title.text
        video_titles.append(video_title)
        print(f"Title {i}: {video_title}")
    
    return video_titles

# Call the function
video_titles = extract_video_titles()


Q4. Python Program to Extract the Number of Views of the First Five Videos
import requests
from bs4 import BeautifulSoup

def extract_video_views():
    url = 'https://www.youtube.com/results?search_query=python+tutorial'  # Example URL
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    video_views = []
    for i, meta in enumerate(soup.find_all('ul', class_='yt-lockup-meta-info')[:5], start=1):
        views = meta.find_all('li')[1].text.strip()
        video_views.append(views)
        print(f"Views {i}: {views}")
    
    return video_views

# Call the function
video_views = extract_video_views()


Q5. Python Program to Extract the Time of Posting of Video for the First Five Videos
import requests
from bs4 import BeautifulSoup

def extract_video_post_times():
    url = 'https://www.youtube.com/results?search_query=python+tutorial'  # Example URL
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    post_times = []
    for i, meta in enumerate(soup.find_all('ul', class_='yt-lockup-meta-info')[:5], start=1):
        time = meta.find_all('li')[0].text.strip()
        post_times.append(time)
        print(f"Post Time {i}: {time}")
    
    return post_times

# Call the function
video_post_times = extract_video_post_times()
Explanation:
URL: We're using a sample search URL for YouTube (https://www.youtube.com/results?search_query=python+tutorial) to demonstrate scraping.
Requests and BeautifulSoup: requests fetches the HTML content of the page, and BeautifulSoup parses and navigates through the HTML.
Functions: Each function (extract_video_urls(), extract_thumbnail_urls(), etc.) follows a similar structure:
Fetch the HTML content.
Parse the content using BeautifulSoup.
Extract the desired information using appropriate HTML tags and classes.
Print and return the extracted data.
These scripts will print the URLs, thumbnails, titles, views, and posting times of the first five videos found on the YouTube search page for "python tutorial". Adjust the URL and parsing logic based on the specific website and content structure you are targeting.

```
