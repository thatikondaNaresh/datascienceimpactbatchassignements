```python
Q1. What are the three measures of central tendency?
The three measures of central tendency are:

Mean: The average value of a dataset, calculated as the sum of all values divided by the number of values.
Median: The middle value in a dataset when the values are arranged in ascending or descending order.
Mode: The value that appears most frequently in a dataset.
Q2. What is the difference between the mean, median, and mode? How are they used to measure the central tendency of a dataset?
Mean: Represents the typical value and is sensitive to outliers.
Median: Represents the middle value and is less affected by extreme values (outliers).
Mode: Represents the most frequent value(s) and is used mainly for categorical data or when identifying peak values in distributions.
These measures are used to provide insights into where the center of the data lies and are chosen based on the distribution and characteristics of the dataset.

â€‹
 5. How are measures of dispersion such as range, variance, and standard deviation used to describe the spread of a dataset? Provide an example.
Range: Measures the difference between the maximum and minimum values, providing a simple measure of spread.
Variance: Measures the average squared deviation from the mean, showing how much the data points vary from the mean value.
Standard Deviation: Provides a standardized measure of the amount of variation or dispersion of a set of values.
Example: For a dataset of exam scores, a high standard deviation indicates that scores are spread out over a wider range, whereas a low standard deviation indicates scores are clustered around the mean.

Q6. What is a Venn diagram?
A Venn diagram is a visual representation used to show all possible logical relations between a finite collection of different sets. It uses overlapping circles to illustrate the relationships between sets, showing where elements of different sets overlap or are distinct.

Q8. What do you understand about skewness in data?
Skewness in data refers to the lack of symmetry in a dataset's distribution. It indicates the degree of asymmetry from the normal distribution in a set of statistical data.

Positive Skewness: The right tail of the distribution is longer or fatter than the left tail, indicating that the distribution has a few unusually high values (right-skewed).
Negative Skewness: The left tail of the distribution is longer or fatter than the right tail, indicating that the distribution has a few unusually low values (left-skewed).
Zero Skewness: The data distribution is perfectly symmetrical.
Q9. If a data is right skewed then what will be the position of median with respect to mean?
If a dataset is right-skewed:

The mean will be greater than the median.
The median will be closer to the left (lower) tail of the distribution where most values are concentrated, while the mean is pulled towards the higher values in the right tail.
Q10. Explain the difference between covariance and correlation. How are these measures used in statistical analysis?
Covariance: Measures the extent to which two variables change together. It indicates the direction (positive or negative) of the linear relationship between variables but does not indicate the strength.
Correlation: Standardized measure that indicates both the direction and strength of the linear relationship between two variables. It ranges from -1 to 1, where 1 indicates a perfect positive correlation, -1 indicates a perfect negative correlation, and 0 indicates no linear correlation.
Use in Statistical Analysis:

Covariance: Used to understand the direction of relationship between two variables but is scale-dependent and does not provide a standardized measure.
Correlation: Provides a standardized measure of the strength and direction of the linear relationship, making it easier to interpret and compare across different datasets.
Q12. For a normal distribution data what is the relationship between its measure of central tendency?
For data that follows a normal distribution:

The mean, median, and mode are all equal.
This symmetry around the mean (average) indicates that data points are evenly distributed around the central value.
Q13. How is covariance different from correlation?
Covariance: Measures the direction and magnitude of the linear relationship between two variables. It is scale-dependent and can take any value.
Correlation: Standardized measure that indicates both the direction and strength of the linear relationship between two variables. It ranges from -1 to 1, where 1 indicates a perfect positive correlation, -1 indicates a perfect negative correlation, and 0 indicates no linear correlation.
Q14. How do outliers affect measures of central tendency and dispersion? Provide an example.
Outliers are data points that lie far outside the usual range of values in a dataset. They can significantly affect measures of central tendency and dispersion:

Central Tendency: Outliers can pull the mean towards them, making it less representative of the typical values in the dataset. The median, being resistant to outliers, is less affected.
Dispersion: Outliers can increase the range, variance, and standard deviation of the dataset, as these measures are sensitive to extreme values.
Example: Consider a dataset of salaries where most employees earn between $40,000 and $60,000 per year, but one executive earns $1,000,000. This outlier significantly increases the mean salary while leaving the median relatively unchanged, indicating its impact on measures of central tendency. Similarly, the range, variance, and standard deviation would also be inflated by this outlier.
```
