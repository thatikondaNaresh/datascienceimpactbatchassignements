```python
Q1. Opening a File and File Modes
Function to Open a File:
The function used to open a file is open().
Modes of Opening a File:
'r': Read mode (default mode). Opens the file for reading. If the file does not exist, it raises a FileNotFoundError.
'w': Write mode. Opens the file for writing. If the file already exists, it truncates the file. If the file does not exist, it creates a new file.
'x': Exclusive creation mode. Creates a new file. If the file already exists, it raises a FileExistsError.
'a': Append mode. Opens the file for writing. If the file already exists, it appends new data to the end of the file. If the file does not exist, it creates a new file.
'b': Binary mode. Opens the file in binary mode. This mode is used for non-text files (e.g., images).
't': Text mode (default mode). Opens the file in text mode.
'+': Update mode. Opens the file for both reading and writing.
Examples:
'r' - Read text mode
'rb' - Read binary mode
'w' - Write text mode
'wb' - Write binary mode
'a' - Append text mode
'ab' - Append binary mode

Q2. Importance of close() Function
The close() function is used to close an open file. It is important to close a file to:
Free up system resources that were being used by the file.
Ensure that data is properly written and saved to the file.
Avoid potential file corruption or data loss.
Allow other processes or programs to access the file.

Q3. Creating, Writing, and Reading a File
# Create and write to a text file
file = open("data_scientist.txt", "w")
file.write("I want to become a Data Scientist")
file.close()

# Open and read the file
file = open("data_scientist.txt", "r")
content = file.read()
print(content)
file.close()

Q4. read(), readline(), and readlines() Functions
read(): Reads the entire content of the file as a single string.
with open("data_scientist.txt", "r") as file:
    content = file.read()
    print(content)
readline(): Reads one line from the file at a time.
with open("data_scientist.txt", "r") as file:
    line = file.readline()
    print(line)
readlines(): Reads all lines of the file and returns them as a list of strings.
with open("data_scientist.txt", "r") as file:
    lines = file.readlines()
    print(lines)

Q5. Using with Statement with open()
The with statement is used to wrap the execution of a block of code and ensure that resources are properly managed. When used with open(), it automatically closes the file when the block of code is exited, even if an error occurs.

Advantages:

Ensures the file is always closed properly.
Reduces the risk of file corruption or data loss.
Simplifies the code by removing the need to explicitly close the file.
Example:
with open("data_scientist.txt", "r") as file:
    content = file.read()
    print(content)
# No need to call file.close(), it's done automatically

Q6. write() and writelines() Functions
write(): Writes a string to the file.
with open("data_scientist.txt", "w") as file:
    file.write("I want to become a Data Scientist")
writelines(): Writes a list of strings to the file.
lines = ["I want to become a Data Scientist\n", "I am learning Python\n", "I love data analysis\n"]
with open("data_scientist.txt", "w") as file:
    file.writelines(lines)
Example:
# Using write()
with open("example.txt", "w") as file:
    file.write("This is a single line written with write().\n")

# Using writelines()
lines = ["This is the first line.\n", "This is the second line.\n", "This is the third line.\n"]
with open("example.txt", "a") as file:
    file.writelines(lines)
In the above example, write() is used to write a single string, while writelines() is used to write multiple lines from a list of strings.
```
